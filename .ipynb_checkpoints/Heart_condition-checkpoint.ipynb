### 1. Importing modules

```python
#Importing basics modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split


#Importing modules necessary to create and validate prediction models

from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline

import warnings
warnings.filterwarnings("ignore")
```

### 2. Importing data

```python
data=pd.read_csv('heart.csv',
                sep=',')


col_names_X=[x for x in data.columns if x!='HeartDisease']
col_name_y=['HeartDisease']
X=data[col_names_X]
y=data[col_name_y]
```

```python
data.head()
```

### 3. Creating basic functions

```python
def data_plots(X,col_names_X):
    """
    The purpose of data_plots is to visualize
    data used in this project
    """
    figure,axis= plt.subplots(3,4,figsize=(14,10))
    for n1 in range(0,4):
        if n1==0:
            for n2,col in enumerate(col_names_X[0:4]):
                axis[n1,n2].scatter(X.index,X[col])
                axis[n1,n2].set_title(col)
        if n1==1:
            for n2,col in enumerate(col_names_X[4:8]):
                axis[n1,n2].scatter(X.index,X[col])
                axis[n1,n2].set_title(col)
        if n1==2:
            for n2,col in enumerate(col_names_X[8:]):
                axis[n1,n2].scatter(X.index,X[col])
                axis[n1,n2].set_title(col)

def data_configuration(X,col_names_X):
    """
    The puprose of data_configuration is to
    change data in columns where there are less
    than 10 unique values to integer values.
    """
    X_new=X
    for col in col_names_X:
        val_counts=len(X_new[col].value_counts())
        if val_counts<10:
            answers=[x for x in X_new[col].value_counts().index]
            dicto={}
            for number,answers in enumerate(answers):
                dicto[answers]=number
            X_new[col]=X_new[col].map(dicto)
        else:
            None
    return X_new  
```

### 4. Data visualization

```python
# data plots before configuration
data_plots(X,col_names_X)
```

```python
#Changing string to integer data
X=data_configuration(X,col_names_X)

"""
Changing the data in cholesterol column
with 0 values for the mean in this data segment
"""
X['Cholesterol']=X['Cholesterol'].replace(0,X['Cholesterol'].mean())

"""
Deleting people with almost zero blood pressure
"""
y=y[X['RestingBP']>20]
X=X[X['RestingBP']>20]
```

```python
"""
Data plots after configuration
"""
data_plots(X,col_names_X)
```

```python
#Correlation visualization between individual columns
corr=X.corr()
sns.heatmap(corr)
plt.title('Data correlation')
plt.show()
```

### 5. Creating prediction models

```python
# Data standarization
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
# Dividing the dataset into test and train data.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
def logistic_regression(model,X_train,y_train,X_test,y_test):
    '''
    The purpose of logistic_regression function is to choose
    the most optimal parameters used in logistic regression
    
    The function returns the model and the score of accuracy 
    score made with test data
    '''
    pipeline=make_pipeline(model)
    grid=GridSearchCV(pipeline,
                      [{'logisticregression__penalty':['none']},
                        {'logisticregression__penalty':['l1'],
                        'logisticregression__C':[0.001,0.01,0.1,1,5,9,10,100]},
                         {'logisticregression__penalty':['l2'],
                        'logisticregression__C':[0.001,0.01,0.1,1,5,9,
                                                 10,100]}
                        ],
                 cv=10)
    grid.fit(X_train,y_train)
    return (model, accuracy_score(grid.best_estimator_.predict(X_test),y_test))
    
def decision_tree_classifier(model,X_train,y_train,X_test,y_test):
    '''
    The purpose of decision_tree_classifier function is to choose
    the most optimal parameters used in logistic regression
    
    The function returns the model and the score of accuracy 
    score made with test data
    '''
    grid=GridSearchCV(model,
                      param_grid={'min_samples_leaf':[1,10,25,50,100]},
                     n_jobs=1)
    grid.fit(X_train,y_train)
    return (model, accuracy_score(grid.best_estimator_.predict(X_test),y_test))

def bernoulli(model,X_train,y_train,X_test,y_test):
    '''
    The purpose of bernoulli function is to return the model
    with the score of accuracy score made with test data
    '''
    model.fit(X_train,y_train)
    return (model, accuracy_score(model.predict(X_test),y_test))
    
def best_model(modele,X_train,y_train,X_test,y_test):
    '''
    The purpose of best_model function is to choose the best
    prediction model using: logistic_regression, decision_tree_classifier
    and bernoulli functions
    This funtion return the best model, best accuracy score and data frame
    with all models used in this project with their accuracy scores made
    by test data
    '''
    results_of_models={}
    best_model=None
    best_as=0
    for number in range(len(modele)):
        if number==0:
            model0,result0=logistic_regression(modele[number],X_train,y_train,X_test,y_test)
            results_of_models[model0]=[result0]
            if result0>best_as:
                best_as=result0
                best_model=model0
        
        if number==1:
            model1,result1=decision_tree_classifier(modele[number],X_train,y_train,X_test,y_test)
            results_of_models[model1]=[result1]
            if result1>best_as:
                best_as=result1
                best_model=model1
        if number==2:
            model2,result2=bernoulli(modele[number],X_train,y_train,X_test,y_test)
            results_of_models[model2]=[result2]
            if result2>best_as:
                best_as=result0
                best_model=model0
    results_of_models=pd.DataFrame.from_dict(results_of_models)
            
    return (best_model, best_as,results_of_models)       
```

### 6.Results

```python
# results of function from creating prediction models
models=[LogisticRegression(),DecisionTreeClassifier(),BernoulliNB()]
best_model,best_score,model_results=best_model(models,X_train,y_train,X_test,y_test)
```

```python
naive_method=y_train['HeartDisease'].value_counts()[1]/\
(y_train['HeartDisease'].value_counts()[0]+\
 y_train['HeartDisease'].value_counts()[0])
model_results['naive_method']=naive_method
```
